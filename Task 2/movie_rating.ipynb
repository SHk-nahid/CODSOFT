```python
import pandas as pd
df = pd.read_csv("movie_rating.csv",  encoding='latin1')
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df = df.dropna(subset=['Rating'])
```

```python
df['Duration']= df['Duration'].str.replace('min','',regex=False)
df['Duration']= pd.to_numeric(df['Duration'],errors='coerce')

df = df.dropna(subset=['Duration'])
df.head()
```

```python
df = df[['Year', 'Duration', 'Genre', 'Votes', 'Director', 'Actor 1', 'Actor 2', 'Actor 3', 'Rating']]
df.head()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df['Genre'] = df['Genre'].fillna(df['Genre'].mode()[0])
```

```python
df.loc[:, 'Genre'] = df['Genre'].fillna(df['Genre'].mode()[0])
df.loc[:, 'Director'] = df['Director'].fillna(df['Director'].mode()[0])
df.loc[:, 'Actor 1'] = df['Actor 1'].fillna('Unknown')
df.loc[:, 'Actor 2'] = df['Actor 2'].fillna('Unknown')
df.loc[:, 'Actor 3'] = df['Actor 3'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df['Rating'], bins=20, kde=True)
plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(y='Genre', data=df, order=df['Genre'].value_counts().index)
plt.title("Genre Count")
plt.show()
```

```python
plt.figure(figsize=(12,6))

top_genres = df['Genre'].value_counts().head(10)

sns.barplot(x=top_genres.values, y=top_genres.index)

plt.title("Top 10 Genres")
plt.xlabel("Count")
plt.ylabel("Genre")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Rating'],bins=20, kde=True, color="teal")
plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='Votes',y='Rating',data=df)
plt.title("Votes vs Rating")
plt.xlabel("Votes")
plt.ylabel("Rating")
plt.show()
```

```python
df['Votes'] = df['Votes'].astype(str).str.replace(',', '').astype(float)

```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[['Duration','Votes','Rating']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='Votes', y='Rating', data=df, alpha=0.5)
plt.title("Votes vs Rating")
plt.xlabel("Votes")
plt.ylabel("Rating")
plt.show()
```
